This notebook imports from the `fcmold_analysis` package and runs the
analysis pipeline.  All logic lives in the package modules:

| Module | Contents |
|--------|----------|
| `config.py` | `AnalysisConfig`, `StrandConfig`, strand definitions |

```python
# Ensure the package is importable (Databricks: add to sys.path if needed)
import sys, os
notebook_dir = os.path.dirname(os.path.abspath('__file__'))
if notebook_dir not in sys.path:
    sys.path.insert(0, notebook_dir)

from fcmold_analysis.config import CONFIG, STRAND_CONFIGS, AnalysisConfig
from fcmold_analysis.pipeline import StrandAnalysisPipeline, run_all_strands

print(f"Config: {CONFIG}")
print(f"Strands available: {list(STRAND_CONFIGS.keys())}")
```

## Single-strand analysis

Change `STRAND_TO_ANALYZE` to `"23_5"` or `"23_6"`.

```python
STRAND_TO_ANALYZE = "23_6"

pipeline = StrandAnalysisPipeline(
    strand_config=STRAND_CONFIGS[STRAND_TO_ANALYZE],
    analysis_config=CONFIG,
)
results = pipeline.run(generate_visualizations=True)

if results["success"]:
    print(f"\nRaw rows: {results['raw_row_count']:,}")
    print(f"Sequences: {results['sequence_count']}")
else:
    print(f"FAILED: {results['error']}")
```
